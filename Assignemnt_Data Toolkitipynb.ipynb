# Data Toolkit

# 1. What is NumPy, and why is it widely used in Python?
  - NumPy is a Python library for handling numerical computations  efficiently. It provides a powerful array object called ndarray, which is faster and more memory-efficient than regular Python lists. Scientists and engineers use it a lot because it makes operations on large datasets smooth and easy, especially matrix math, linear algebra, and element-wise calculations.

# 2.How does broadcasting work in NumPy?
   - Broadcasting is like magic that allows NumPy to perform operations on arrays of different shapes without explicitly reshaping them. For example, if you add a 1D array to a 2D array, NumPy automatically stretches the smaller array along the missing dimensions so the shapes match, making operations very concise.

# 3. What is a Pandas DataFrame?
  - A DataFrame is like an Excel spreadsheet inside Python. It’s a 2D labeled data structure where each column can have a different type. You can access rows and columns easily, perform calculations, filter data, or group it. Think of it as a supercharged table that lets you manipulate and analyze data efficiently.

# 4.Explain the use of the groupby() method in Pandas.
  - groupby() is used to split your data into groups based on some criteria and then perform operations like sum, mean, or custom functions on each group. For example, you could group sales data by region and quickly find total sales per region. It’s like categorizing your data and summarizing it automatically.

# 5. Why is Seaborn preferred for statistical visualizations?
  - Seaborn is designed to make statistical plots look good with minimal code. It works beautifully with Pandas DataFrames, has built-in support for themes, colors, and aggregations, and can handle complex plots like violin plots, boxplots, and regression plots easily. Basically, it makes data look elegant and meaningful without much hassle.

# 6.  What are the differences between NumPy arrays and Python lists?
-   Speed & efficiency: NumPy arrays are faster and take less memory.
 - Operations: Arrays support vectorized math, meaning you can add, multiply, or apply functions to the whole array without loops.

 - Shape: Arrays can be multidimensional; lists are mostly 1D (though you can nest them).

# 7. What is a heatmap, and when should it be used?
   - A heatmap is a visual grid where values are represented by colors. It’s great for showing correlations, confusion matrices, or any 2D data where you want to see patterns or intensity quickly.

# 8. A What does the term “vectorized operation” mean in NumPy?
- It means performing operations on entire arrays at once instead of looping through elements. For example, arr + 5 adds 5 to every element without a loop. It’s faster and cleaner than traditional Python loops.

# 9. How does Matplotlib differ from Plotly?
   - Matplotlib is a static plotting library — it creates traditional charts like line, bar, and scatter plots. Plotly, on the other hand, is interactive — you can zoom, hover, and pan, and it’s ideal for dashboards and web-based visualizations.

# 10.  What is the significance of hierarchical indexing in Pandas?
   - Hierarchical indexing (MultiIndex) allows you to index rows and columns with multiple levels, like nesting categories inside each other. It makes complex datasets easier to slice and analyze without flattening the data.


# 11. What is the role of Seaborn’s pairplot() function?
   - pairplot() shows scatterplots for every pair of numeric columns and histograms on the diagonal. It’s useful for quickly exploring relationships and spotting trends or correlations in your dataset.

# 12. What is the purpose of the describe() function in Pandas?
   - describe() gives you a quick statistical summary of numeric columns — mean, standard deviation, min, max, and quartiles. It’s like a first glance at your data to understand its spread and central tendencies.

# 13. Why is handling missing data important in Pandas?
  - Missing data can break calculations, distort statistics, and affect machine learning models. Handling them properly — by filling, dropping, or imputing — ensures your analysis is accurate and reliable.

# 14. What are the benefits of using Plotly for data visualization?
   - Plotly creates interactive, web-ready visualizations. You can zoom, hover, pan, and even embed plots in dashboards. It’s perfect when you want to explore data dynamically or present it online.

# 15. A How does NumPy handle multidimensional arrays?
   - NumPy arrays can have any number of dimensions (1D, 2D, 3D, etc.). You can slice, reshape, and perform operations along any axis. This is crucial for images, matrices, and scientific data where data isn’t just flat.

# 16. What is the role of Bokeh in data visualization?
   - Bokeh is a Python library for interactive web visualizations. It’s similar to Plotly but gives you fine-grained control over plots and integrates well into web apps and dashboards.

# 17. Explain the difference between apply() and map() in Pandas.
   - map() → applies a function to each element of a Series.

- apply() → can apply a function to an entire Series or DataFrame, row-wise or column-wise.

  Think of map() for small element-wise transformations, apply() for bigger, custom operations.

# 18.What are some advanced features of NumPy?
   - Broadcasting for different-shaped arrays.

- Vectorized math for speed.

-  Multidimensional arrays and advanced indexing.

- Linear algebra, random sampling, FFTs.

- Integration with C/C++ code for speed.

 # 19. How does Pandas simplify time series analysis?
  - Pandas has DateTimeIndex, resampling, rolling windows, and shift operations that make it easy to analyze trends, compute moving averages, and align dates. Time series becomes almost effortless compared to raw Python.

# 20. What is the role of a pivot table in Pandas?
   - Pivot tables reshape and summarize data. You can aggregate values (like sum or mean) across categories, creating a table that’s easy to analyze, similar to Excel pivot tables

 # 21. Why is NumPy’s array slicing faster than Python’s list slicing?
  - NumPy arrays are stored in contiguous blocks of memory, unlike Python lists, which are collections of separate objects. When you slice a NumPy array, it creates a view of the original data rather than copying it, so no new memory allocation is needed. Python lists, on the other hand, create a new list when sliced, which involves copying elements. This makes NumPy slicing much faster and more memory-efficient, especially for large datasets.

# 22. What are some common use cases for Seaborn?
   Seaborn is perfect for statistical and exploratory data analysis. Common use cases include:

- Visualizing distributions with histograms, KDE plots, and violin plots.

- Comparing categories using boxplots, barplots, or swarmplots.

- Exploring relationships with scatterplots, regression plots, and pairplots.

- Showing correlations via heatmaps.

- Quickly making publication-quality plots with minimal styling effort.

```python
# 1.How do you create a 2D NumPy array and calculate the sum of each row?
import numpy as np
arr=np.array([[1,2,3],[4,5,6]])
print(arr)
print("Sum of each row :" ,arr.sum(axis=1))
```

```python
# 2. Write a Pandas script to find the mean of a specific column in a DataFrame.
import pandas as pd
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'Age': [23, 35, 30, 40],
    'Score': [85, 90, 88, 92]
}
df = pd.DataFrame(data)
mean_age = df['Age'].mean()
print("Mean Age:", mean_age)
```

```python
# 3.Create a scatter plot using Matplotlib.
import matplotlib.pyplot as plt

x = [5, 7, 8, 7, 6, 9, 5, 6, 7, 8]
y = [99, 86, 87, 88, 100, 86, 103, 87, 94, 78]
plt.scatter(x,y)
plt.xlabel("x-axis")
plt.ylabel("y-axis")
plt.title("Scatter Plot")
plt.show()
```

```python
# 4. How do you calculate the correlation matrix using Seaborn and visualize it with a heatmap?
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
data={
    'Math': [85, 90, 78, 92, 88],
    'Physics': [80, 85, 75, 89, 84],
    'Chemistry': [82, 88, 79, 91, 86],
    'Biology': [70, 75, 72, 78, 74]
}
df=pd.DataFrame(data)
correlation_matrix = df.corr()
plt.figure(figsize=(6, 4))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title("Correlation Matrix Heatmap")
plt.show()
```

```python
# 5. Generate a bar plot using Plotly.
import plotly.express as px
import pandas as pd
data={ "Subjects": ["Math", "Physics", "Chemistry", "Biology"],
    "Scores": [85, 90, 78, 88]}
df=pd.DataFrame(data)
fig = px.bar(df,
             x="Subjects",
             y="Scores",
             title="Student Scores by Subject",
             color="Subjects")
fig.show()
```

```python
# 6. Create a DataFrame and add a new column based on an existing column.
import pandas as pd
data={
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'Score': [85, 90, 78, 92]
    }
df=pd.DataFrame(data)
df['Grade'] = df['Score'].apply(lambda x: 'A' if x >= 90 else ('B' if x >= 80 else 'C'))
print(df)
```

```python
# 7. Write a program to perform element-wise multiplication of two NumPy arrays.
import numpy as np
arr1 = np.array([1, 2, 3, 4, 5])
arr2 = np.array([10, 20, 30, 40, 50])
result = arr1 * arr2
print("Array 1:", arr1)
print("Array 2:", arr2)
print("Element-wise multiplication:", result)
```

```python
# 8. Create a line plot with multiple lines using Matplotlib.
import matplotlib.pyplot as plt
import numpy as np
x = [1, 2, 3, 4, 5]
y1 = [2, 4, 6, 8, 10]
y2 = [1, 3, 5, 7, 9]
y3 = [2, 3, 4, 5, 6]
plt.plot(x, y1, label='Line 1')
plt.plot(x, y2, label='Line 2')
plt.plot(x, y3, label='Line 3')
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Multiple Line Plot')
plt.legend()
plt.show()
```

```python
# 9. Generate a Pandas DataFrame and filter rows where a column value is greater than a threshold.
import pandas as pd
data={
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'Score': [85, 90, 78, 92]
    }
df=pd.DataFrame(data)
threshold = 80
filtered_df = df[df['Score'] > threshold]
print("Original DataFrame:")
print(df)
print("\nDataFrame after filtering:")
print(filtered_df)
```

```python
# 10. Create a histogram using Seaborn to visualize a distribution.
import seaborn as sns
import matplotlib.pyplot as plt
data = np.random.randint(50, 101, size=100)
sns.histplot(data, kde=True)
plt.xlabel('Value')
plt.ylabel('Frequency')
plt.title('Histogram')
plt.show()
```

```python
# 11. Perform matrix multiplication using NumPy.
import numpy as np
matrix1 = np.array([[1, 2], [3, 4]])
matrix2 = np.array([[5, 6], [7, 8]])
result = np.dot(matrix1, matrix2)
print("Matrix 1:")
print(matrix1)
print("\nMatrix 2:")
print(matrix2)
print("\nMatrix Multiplication Result:")
print(result)
```
